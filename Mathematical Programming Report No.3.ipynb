# 練習問題1
## 問題
$$ Maximize\\ Z = X_1 + X_2 + X_3 \\ Subject\ to \\ X_1 - X_2 + 2X_3 = 8 \\ 2X_1 - 3X_2 - X_3 = 1 \\ X_j \geq 0, j=1,2,3$$

## 回答
今回の問題について制限標準型に変形を行うと以下のようになる．
$$ Minimize\\ Z = -X_1 - X_2 - X_3 \\ Subject\ to \\ X_1 - X_2 + 2X_3 + X_4 = 8 \\ 2X_1 - 3X_2 - X_3 + X_5 = 1 \\ X_j \geq 0, j=1,2,3,4,5$$


```python
import pandas as pd
import numpy as np

epoch = 5

df = pd.DataFrame({ 'Z' : [-1,-1,-1,0],
                                    'X4': [1,-1,2,8],
                                    'X5': [2,-3,-1,1]},
                                    index = ['X1','X2','X3','C'])
df = df.T # 行列の転置
df
```

```python
def simplex_method(df):
    for i in range(epoch):
        print("epoch {}".format(i+1))
        # select pivot
        pivot_col = df.loc['Z'].drop('C').argmin()
        pivot_ind = df[pivot_col].drop('Z').argmax()
        pivot_num = max(df[pivot_col])

        # divide select index
        df.loc[pivot_ind] /= pivot_num
        print("行選択とpivot計算")
        df.rename(columns={pivot_col:pivot_ind},index={pivot_ind:pivot_col},inplace=True)
        pivot_ind,pivot_col = pivot_col,pivot_ind
        df_ans = df.copy()
        display(df)

        ## Sweeping-out method
        # remove pivot index & columns
        df_piv_col = df.loc[pivot_ind].drop(pivot_col)
        df_piv_ind = df[pivot_col].drop(pivot_ind)
        sweep_df = df.drop(pivot_col, axis=1, inplace=True)       
        sweep_df = df.drop(pivot_ind)     

        # Sweeping
        mat_x1 = np.array([df_piv_ind.T for i in range(len(df_piv_col))]).T
        mat_x2 = np.diag(df_piv_col)
        mat_y = -np.dot(mat_x1,mat_x2)

        sweep_df += mat_y
        df_ans = pd.concat([df_ans,sweep_df]).fillna(0)
        
        # old index deleted
        grouped = df_ans.groupby(level=0)
        df_ans = grouped.last()

        print("掃き出し法後")
        display(df_ans)
        
        # judge optimizied...
        if sum(df_ans.loc['Z'] < 0) == 0:
            break

        df = df_ans
```

```python
simplex_method(df)
```
